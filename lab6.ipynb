### One-Hot Encoding, Label Encoding, Frequency Encoding, PCA, RFE

### Encoding Categorical Variables

```python
import pandas as pd
import numpy as np

data = {'Color': ['Red', 'Blue', 'Green', 'Blue', 'Red']}
df = pd.DataFrame(data)

# One-Hot Encoding
encoded_df = pd.get_dummies(df, columns=['Color'], drop_first=False)
print("One-Hot Encoded:\n", encoded_df)

# Label Encoding
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['Color_Label'] = label_encoder.fit_transform(df['Color'])
print("\nLabel Encoded:\n", df)

# Frequency Encoding
frequency_map = df['Color'].value_counts().to_dict()
df['Color_Frequency'] = df['Color'].map(frequency_map)
print("\nFrequency Encoded:\n", df)
```

### Polynomial Features & Rolling Statistics

```python
from sklearn.preprocessing import PolynomialFeatures

df['Feature1'] = np.random.rand(len(df))
df['Feature2'] = np.random.rand(len(df))

poly = PolynomialFeatures(degree=2, interaction_only=False, include_bias=False)
poly_features = poly.fit_transform(df[['Feature1', 'Feature2']])
print("Polynomial Features:\n", poly_features)

df['Value'] = np.random.rand(len(df))
df['Rolling_Mean'] = df['Value'].rolling(window=3).mean()
df['Rolling_Std'] = df['Value'].rolling(window=3).std()
print("\nWith Rolling Stats:\n", df)
```

### Feature Selection — Correlation, SelectKBest, RFE

```python
from sklearn.feature_selection import f_regression, SelectKBest
from sklearn.impute import SimpleImputer

df['Target'] = np.random.rand(len(df))
numerical_df = df.select_dtypes(include=np.number)
correlation = numerical_df.corr()
print("Correlation with Feature1:\n", correlation['Feature1'])

X = df.drop(columns=['Target', 'Color']).dropna()
y = df['Target'][X.index]
X_new = SelectKBest(f_regression, k=5).fit_transform(X, y)
print("\nSelectKBest output:\n", X_new)
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

model = LinearRegression()
rfe = RFE(model, n_features_to_select=5)
X_rfe = rfe.fit_transform(X, y)
print("RFE Selected Features:", rfe.support_)
```
