# LAB 2: Hypothesis Tests (t-tests, Z-tests)
### Compare population means using sample datasets

## One-Sample t-Test

**Scenario:** A school claims the average score is 75. A sample of 10 students' scores is collected. Test if the sample mean significantly differs from the claimed mean (α = 0.05).

```python

import numpy as np
import scipy.stats as stats

scores = [78, 72, 74, 76, 73, 79, 77, 71, 70, 74]

population_mean = 75
n = len(scores)

sample_mean = np.mean(scores)
sample_std = np.std(scores, ddof=1)

# Calculate t-statistic
t_stat = (sample_mean - population_mean) / (sample_std / np.sqrt(n))

# Degrees of freedom
df = n - 1

# Two-tailed p-value
p_value = 2 * stats.t.sf(abs(t_stat), df)

print(f"Sample Mean: {sample_mean:.2f}")
print(f"t-statistic: {t_stat:.2f}")
print(f"p-value: {p_value:.4f}")

if p_value < 0.05:
  print("Reject the null hypothesis: The sample mean is significantly different.")
else:
  print("Fail to reject the null hypothesis: No significant difference.")
```

## Two-Sample t-Test

**Scenario:** Compare exam scores of two classes to see if there is a significant difference.

```python
import numpy as np
import scipy.stats as stats

 # Data
class_a = [85, 88, 90, 92, 86, 84, 87, 89, 91, 93]
class_b = [78, 80, 82, 79, 81, 77, 83, 76, 75, 80]

# Two-sample t-test
t_stat, p_value = stats.ttest_ind(class_a, class_b, equal_var=False)

# Results
print(f"t-statistic: {t_stat:.2f}")
print(f"p_value: {p_value:.4f}")

if p_value < 0.05:
  print("Reject the null hypothesis: The means are significantly different.")
else:
  print("Fail to reject the null hypothesis: No significant difference.")
```

## One-Sample Z-Test

**Scenario:** A factory claims average product weight is 500g. Sample of 40 products has mean 495g, population std = 10g.

```python
import numpy as np
import scipy.stats as stats

# Data
sample_mean = 495
population_mean = 500
population_std = 10
n = 40

# Calculate z-statistic
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# Calculate p-value (two-tailed)
p_value = 2 * stats.norm.sf(abs(z_stat))

# Results
print(f"z-statistic: {z_stat:.2f}")
print(f"p-value: {p_value:.4f}")

if p_value < 0.05:
  print("Reject the null hypothesis: The sample mean is significantly different.")
else:
  print("Fail to reject the null hypothesis: No significant difference.")
```
